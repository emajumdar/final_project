```python
# create subset of dataframe containing only agency, demographic data, and EEI questions
fevs_eei = fevs[['AGENCY', 'DSEX', 'DEDUC', 'DFEDTEN', 'DSUPER', 'DMINORITY', 'DLEAVING', 'Q3', 'Q4', 'Q6', 'Q11', 'Q12', 'Q47', 'Q48', 'Q49', 'Q51', 'Q52', 'Q53', 'Q54', 'Q56', 'Q60', 'Q61']]
```

```python
# function to calculate total positive responses (ratings 4 and 5) for each column
def ratings(column):
    rate_4 = (fevs_eei[column] == 4).sum()
    rate_5 = (fevs_eei[column] == 5).sum()
    total_positive = rate_4 + rate_5
    return total_positive
```

```python
# test
ratings('Q3')
```
