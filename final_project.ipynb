## Logistic Regression

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(font_scale=1.5);
```

```python
#Scatter plot with regression line
sns.lmplot(x='total_eei', y='DLEAVING', data=fevs_eei);
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
```

```python
# Set X and y
feature_cols = ['total_eei']
X = fevs_eei[feature_cols]
y = fevs_eei.DLEAVING
```

```python
# Fit model
logreg.fit(X,y)
pred = logreg.predict(X)
```

```python
# Plot the class predictions.
plt.figure(figsize=(10,10))
plt.scatter(fevs_eei.sort_values(by='total_eei').total_eei, fevs_eei.DLEAVING)
plt.plot(fevs_eei.sort_values(by='total_eei').total_eei, pred, color='red')
plt.xlabel('Employee Engagement Score')
plt.ylabel('Intent to Leave')
```

```python
# Compute predicted probabilities.
fevs_eei['Turnover_Prob'] = logreg.predict_proba(X)[:, 1]
```

```python
# Resample the data to prevent an overflow error when creating visualization
sample = fevs_eei.sample(n=10000)
```

```python
sample.head()
```

```python
# Plot the predicted probabilities based on the sampled data.
plt.figure(figsize=(10,10))
plt.scatter(sample.total_eei, sample.DLEAVING)
plt.plot(sample.sort_values(by='total_eei').total_eei, sample.sort_values(by='total_eei').Turnover_Prob, color='red')
plt.xlabel('Employee Engagement Score', fontsize = 20)
plt.ylabel('Intent to Leave', fontsize = 20)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
X = fevs_eei[['total_eei']]
y = fevs_eei.DLEAVING

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y)

LR.fit(X_train,y_train)

y_pred = LR.predict(X_test)

LR.score(X_test, y_test)
```

```python
# Find y_pred
y_pred = LR.predict(fevs_eei[['total_eei']])
```

```python
# Compute predicted log odds for total_eei=50 using the equation.
logodds = logreg.intercept_ + logreg.coef_[0] * 50
logodds
```

```python
# Convert log odds to odds.
odds = np.exp(logodds)
odds
```

```python
# Convert odds to probability.
# When the EEI score is 50, there is a 28% chance the employee intends to leave. 
prob = odds/(1 + odds)
prob
```

```python
# Confirm this is what the model would produce.
logreg.predict_proba(50)
```

```python
# Examine the coefficient for total_eei.
# A 1-unit increase in total_eei is associated with a 0.07-unit decrease in the log odds of intent to leave.
(feature_cols, logreg.coef_[0])
```
